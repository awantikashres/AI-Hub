# 3.3 Exploratory Data Analysis with Group-by Method Practice Exercise:

What percent of the total fare revenue came from each passenger class?
To answer the question perform following operation:
1. Calculate the total Fare paid across all passengers in the Titanic dataset.

```python
total_fare=titanic_data['Fare'].sum()
print(total_fare)
```

2. Subset for passengers in first class (Pclass is 1) and calculate their total fare.

```python
first_class_pessengers = titanic_data[titanic_data['Pclass']==1]
total_fare_first_class = first_class_pessengers['Fare'].sum()
print("Total fare for first-class passengers: ", total_fare_first_class)
```

3. Do the same for second class (Pclass is 2) and third class (Pclass is 3).

```python
second_class_passengers = titanic_data[titanic_data['Pclass']==2]
total_fare_second_class = second_class_passengers['Fare'].sum()
print("Total fare for second-class passengers: ", total_fare_second_class)
```

```python
third_class_passengers = titanic_data[titanic_data['Pclass']==3]
total_fare_third_class = third_class_passengers['Fare'].sum()
print("Total fare for third-class passengers: ", total_fare_third_class)
```

4. Combine the fare totals from first, second, and third classes into a list.

```python
total_fare_per_class = titanic_data.groupby('Pclass')['Fare'].sum()
fare_totals_list = total_fare_per_class.tolist()
print("fare totals by classes: ", fare_totals_list)
```

5. Divide the totals for each class by the overall total fare to get the proportion of fare revenue by class.

```python
total_fare = titanic_data['Fare'].sum()
fare_proportion_per_class = total_fare_per_class / total_fare
fare_proportion_list = fare_proportion_per_class.tolist()
print("Proportion of fare revenue by class: ", fare_proportion_list)
```

What percent of the total number of passengers on the Titanic belonged to each age group
(e.g., child, adult, senior)?
To answer the question perform following operation:
1. Create a new column, age group, that categorizes passengers into "child" (age < 18), "adult" (age
18{64), and "senior" (age 65 and above).

```python
#Create new column
def categorize_age(age):
  if age<18:
    return 'child'
  elif age>=18 and age<65:
    return 'adult'
  else:
    return 'senior'

titanic_data['Age_group']=titanic_data['Age'].apply(categorize_age)
print(titanic_data[['Name', 'Age', 'Age_group']].head())
```

2. Calculate the total number of passengers on the Titanic.

```python
total_passengers=len(titanic_data)
print("Total number of passengers: ", total_passengers)
```

3. Count the number of passengers in each age group.

```python
age_group_counts=titanic_data['Age_group'].value_counts()
print("Number of passengers in each age group: ", age_group_counts)
```

4. Divide the count of each age group by the total number of passengers to get the proportion of passengers
in each age group.

```python
age_group_proportions=age_group_counts/total_passengers
print("Proportion of passengers in each age group: ", age_group_proportions)
```

5. Display the proportion as a percentage.

```python
age_group_percentage = (age_group_counts/total_passengers)*100
print("Percentage of passengers in each age group: ", age_group_percentage)
```
