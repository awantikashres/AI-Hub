# **Problem 4: Matrix Operations**

*For the following arrays:
A = np.array([[3,4],[7,8]]) and B = np.array([[5,3],[2,1]]);
Prove following with Numpy:
1. Prove A.A−1 = I.
2. Prove AB ̸= BA.
3. Prove (AB)

T = BTAT*

*Prove: A.A-1 = I*

```python
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])
A_inv = np.linalg.inv(A)
I = np.dot(A, A_inv)
print(I)
```

*Prove that AB/=BA*

```python
AB = A @ B
BA = B @ A
print("AB = \n", AB)
print("BA = \n", BA)
```

*Prove that (AB)^T = B^T.A^T*

```python
left = (A@B).T
right = B.T @ A.T
print("Left = \n", left)
print("Right = \n", right)
```

*Solve the following system of Linear equation using Inverse Methods.

2x − 3y + z = −1
x − y + 2z = −3
3x + y − z = 9

{Hint: First use Numpy array to represent the equation in Matrix form. Then Solve for: AX = B}*

```python
A = np.array ([
    [2, -3, 1],
    [1, -1, 2],
    [3, 1, -1]
])
B = np.array([[-1],
              [-3],
              [9]])
A_inv = np.linalg.inv(A)
X = A_inv @ B
print(X)
```
