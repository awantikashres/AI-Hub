```python
#Function to classify HDI
def classify_hdi(hdi):
  if hdi<0.550:
    return 'Low'
  elif 0.550 <= hdi <= 0.699:
    return 'Medium'
  elif 0.700 <= hdi <=0.799:
    return 'High'
  else:
    return 'Very High'
```

3. Outlier Detection:

â€¢ Detect outliers in HDI and GNI per Capita using the 1.5 Ã— IQR rule.

â€¢ Create a scatter plot of GNI per Capita vs HDI, highlighting the outliers in a different color.

â€¢ Discuss why the identified countries stand out as outliers.

```python
#Detect outliers
def detect_outliers(column):
  Q1 = column.quantile(0.25)
  Q3 = column.quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 -1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  outliers = column[(column < lower_bound) | (column > upper_bound)]
  return outliers

hdi_outliers = detect_outliers(hdi_problem1B_df['hdi'])
print("HDI Outliers:\n", hdi_outliers)

gni_outliers = detect_outliers(hdi_problem1B_df['gross_inc_percap'])
print("GNI per Capita Outliers:\n", gni_outliers)
```

```python
#Create scatter plot
hdi_problem1B_df['hdi_outlier'] = hdi_problem1B_df['hdi'].isin(hdi_outliers)
hdi_problem1B_df['gni_outlier'] = hdi_problem1B_df['gross_inc_percap'].isin(gni_outliers)
plt.figure(figsize=(10,6))

#Normal points
normal = hdi_problem1B_df[~(hdi_problem1B_df['hdi_outlier'] | hdi_problem1B_df['gni_outlier'])]
plt.scatter(normal['gross_inc_percap'], normal['hdi'], color='blue', label='Normal')

#Outliers
outliers = hdi_problem1B_df[hdi_problem1B_df['hdi_outlier'] | hdi_problem1B_df['gni_outlier']]
plt.scatter(outliers['gross_inc_percap'], outliers['hdi'], color='red', label='Outliers')

plt.xlabel('GNI per Capita')
plt.ylabel('HDI')
plt.title('Scatter Plot of GNI per Capita vs HDI with Outliers Highlighted')
plt.legend()
plt.show()
```

#Discuss for report
Outliers stand out because:

HDI Outliers: A country might have an unusually high or low HDI compared to peers.

Example: High HDI but low GNI could indicate strong social services despite lower income.

GNI per Capita Outliers: Countries with exceptionally high or low income.

Example: Small wealthy countries may appear as GNI outliers.

Overall: Outliers often highlight economic disparities, social development anomalies, or unusual policy outcomes within the South Asian region.

5. HDI Disparity:

â€¢ Compute the range (max â€“ min) of HDI for each region.

â€¢ Compute the coefficient of variation (CV = std/mean) for HDI.

â€¢ Identify which region exhibits more variation in HDI.

```python
#Identify Region with More Variation
if sa_CV > me_CV:
    print("\nSouth Asia exhibits more relative variation in HDI (higher CV).")
else:
    print("\nMiddle East exhibits more relative variation in HDI (higher CV).")
```

7. Outlier Detection:

â€¢ Detect outliers in HDI {"hdi"} and GNI per Capita {"gross inc percap"} for each region using
the 1.5 Ã— IQR rule.

â€¢ Create scatter plots highlighting outliers in a different color.

â€¢ Discuss the significance of these outliers.

```python
#Define Outlier Detection Function
def detect_outliers_iqr(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return series[(series < lower) | (series > upper)]
```

```python
#Detect Outliers in each region
# South Asia
sa_HDI_outliers = detect_outliers_iqr(south_asia_df['hdi'])
sa_GNI_outliers = detect_outliers_iqr(south_asia_df['gross_inc_percap'])

print("South Asia HDI Outliers:\n", sa_HDI_outliers)
print("\nSouth Asia GNI Outliers:\n", sa_GNI_outliers)

# Middle East
me_HDI_outliers = detect_outliers_iqr(middle_east_df['hdi'])
me_GNI_outliers = detect_outliers_iqr(middle_east_df['gross_inc_percap'])

print("\nMiddle East HDI Outliers:\n", me_HDI_outliers)
print("\nMiddle East GNI Outliers:\n", me_GNI_outliers)
```

```python
#Scatter plot highlighting  outliers
def plot_outliers(df, region_name):
    # Mark outliers
    df.loc[:, 'hdi_outlier'] = df['hdi'].isin(detect_outliers_iqr(df['hdi']))
    df.loc[:, 'gni_outlier'] = df['gross_inc_percap'].isin(detect_outliers_iqr(df['gross_inc_percap']))

    plt.figure(figsize=(8,6))

    # Normal points
    normal = df[~(df['hdi_outlier'] | df['gni_outlier'])]
    plt.scatter(normal['gross_inc_percap'], normal['hdi'], color='blue', label='Normal')

    # Outliers
    outliers = df[df['hdi_outlier'] | df['gni_outlier']]
    plt.scatter(outliers['gross_inc_percap'], outliers['hdi'], color='red', label='Outliers')

    plt.xlabel('GNI per Capita')
    plt.ylabel('HDI')
    plt.title(f'HDI vs GNI per Capita with Outliers ({region_name})')
    plt.legend()
    plt.show()

# South Asia
plot_outliers(south_asia_df, 'South Asia')

# Middle East
plot_outliers(middle_east_df, 'Middle East')
```
